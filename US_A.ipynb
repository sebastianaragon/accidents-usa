## Start Time


```python
df.Start_Time
```

```python
df.Start_Time=pd.to_datetime(df.Start_Time)
df.Start_Time
```

```python
ax=sns.distplot(df.Start_Time.dt.hour,bins=24,kde=False,norm_hist=True)#norm_hist is to get percentages
ax.set(xlabel='Hour', ylabel='percentage')
ax.set_title('accidents per hour of the day')
```

-A high percentage of accidents occur between 3-5 pm

```python
ax=sns.distplot(df.Start_Time.dt.dayofweek,bins=7,kde=False,norm_hist=True)
ax.set(xlabel='Days', ylabel='percentage')
ax.set_title('accidents per day')
```

##the peak occurs between weekdays being highest friday 


```python
ax=sns.distplot(df.Start_Time.dt.month,bins=12,kde=False,norm_hist=True)#norm_hist is to get percentages
ax.set(xlabel='Month', ylabel='percentage')
ax.set_title('accidents per month ')
```

## highest accidents occurs in december probably for the holiday
